### Split by Year

```python
df = pd.read_csv("D:\LLMTables\Sportset\sportset_combined.csv")
```

```python
unique_years = df['year'].unique()
```

```python
unique_years
```

```python
for year in unique_years:
    df_year = df[df['year'] == year]
    
    file_name = f"sportset_year_{year}_30.csv"
    sample = df_year.sample(30)
    sample.to_csv(file_name, index=False)

    print(f"CSV file created for year {year}: {file_name}")
```

```python
# Get the unique years from the 'year' column
unique_years = df['year'].unique()

# Loop over each unique year
for year in unique_years:
    # Filter the DataFrame for the current year
    df_year = df[df['year'] == year]
    
    # Check if there are at least 30 rows to sample, skip if not
    if len(df_year) >= 60:
        # Randomly sample 30 rows
        sample = df_year.sample(30)

        # Create a CSV file for each sampled subset
        file_name = f"sportset_year_{year}_60_34.csv"
        sample.to_csv(file_name, index=False)

        print(f"CSV file created for year {year}: {file_name}")

        # Remove the sampled rows from the original DataFrame
        df = df.drop(sample.index)
```
