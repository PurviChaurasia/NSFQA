```python
def generate_questions(table):
    # Prompt asking for complex questions
    prompt = f"""
    ------- TASK -----------------
    TASK: You will analyze the given table which contains information about various sport events that took place along with the dates, locations game and player info.

    -------- ROLE ----------------
    ROLE: You are a complex question and answer generator that works for semi structured tabular data. Your role is to analyse the table given to you and generate 5-7 complex question answer pairs based on the guidlines given to you. 
    NOTE it is important to follow the guidlines given and focus on making complex question answer pairs.

    --------- QUESTION GENERATION GUIDELINES ---------
    GUIDELINES:
    Generate complex, deterministic questions from the following table data.The questions should have the following properties - 
    1. INTER CELL RELATIONSHIPS: The questions should have inter cell relationships.
    2. SINGLE ROW RETRIEVAL: The question should be answerable from a single row of the table
    2. INDIRECT INFERENCE: They can either be directly or indirectly answered from the table. 
    (For eg - for an academic dataset, the question can be "How many scholars are from the filed of AI" and while there is no direct mention of AI there is mention of field of AI such as NLP, CV, stc in the summary)
    3. STRUCTURED AND UNSTRCUTURED DATA: They should make use of both the structured and unstructured components of the table. Something like a questions which starts from extracting infromation from a particular row then goes to a summary to extract the next piece of information and then to another row for the final result.
    4. CONCISE AND DETERMINISTIC: The answers to the question should be deterministic in nature and limited to MAXIMUM 1-2 lines. 

    ----------- TABLE DATA ------------------
    Table Data:
    {table}

    ----------- EXAMPLE QUESTIONS -----------

    Some examples of such complex questions are - 
    Q. Who won the game between the Orlando Magic and the Philadelphia 76ers in November 2014?
    REASONING: In order to answer this question the model will first have to look at all entries for month = november and year = 2014 and then, go through the summaries to find the answer.

    Q. What happened to Nerlens Noel during the November 2014 game?
    REASONING: Similar to the above questions, records will be filtered for the month novermber and year 2014, and then summaries will be scanned to find the answer.


    BASED ON ALL THIS INFORMATION, Please provide a list of such complex questions.
    """
    # Define your messages
    messages = [
        {"role": "system", "content": "You are an expert in generating complex questions from tabular data. Your task is to create questions that require analyzing inter-cell relationships within the provided table and can be answered directly or indirectly using the table's content. The questions should leverage both structured and unstructured components of the table data."},
        {"role": "user", "content": prompt}
    ]

    # Make the API call
    completion = client.chat.completions.create(
        model="gpt-4o-mini",  # or the model you are using
        messages=messages
    )

    # Print the response
    print(completion.choices[0].message.content)
```
